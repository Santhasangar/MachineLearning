```python
import pandas as pd

df= pd.read_csv("50_Startups.csv")
df
```

```python
data= pd.get_dummies(df, dtype=int)
data
```

```python
data.columns
```

```python
X = data[['R&D Spend', 'Administration', 'Marketing Spend', 'State_California', 'State_Florida']]
Y = data[['Profit']]
```

```python
print(X)
```

```python
print(Y)
```

```python
X.columns
```

```python
Y.columns
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plt.scatter(X[['R&D Spend']], Y, label="R&D Spend")
plt.scatter(X[['Administration']],Y, label="Admin")
plt.scatter(X[['Marketing Spend']],Y, label="Marketing")

plt.title("Business cost and Profit Analysis")
plt.xlabel("All business Cost", fontsize=20)
plt.ylabel("Profit", fontsize=20)
plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.30,random_state=0)
```

```python
x_train
```

```python
from sklearn.linear_model import LinearRegression
regression = LinearRegression()
regression.fit(x_train, y_train)
```

```python
y_predict = regression.predict(x_test)
```

```python
y_predict
```

```python
weight = regression.coef_
bias = regression.intercept_
print("Weight of the Model(M or Slope):", weight)
print("Bias of the Model(C or Intercept):", bias)
```

```python
from sklearn.metrics import r2_score
score = r2_score(y_predict, y_test)
score
```

```python
import pickle
filename="MultipleRegression.sav"
pickle.dump(regression, open(filename, "wb"))
```

```python
load_model = pickle.load(open(filename,"rb"))
result = load_model.predict([[1234,345,4565,1,0]],)
result

#R&D Spend	Administration	Marketing Spend	State_California	State_Florida
```
