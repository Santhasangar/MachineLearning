```python
import pickle
filename = "MultipleRegression.sav"
model = pickle.load(open(filename,"rb"))

#Collect input value from users
RD = float(input("Enter value for R&D Cost"))
Admin =float(input("Enter value for Administration Cost"))
Market =float(input("Enter value for Marketing Spend cost"))
California=int(input("Enter true or fale as 0 or 1 for California"))
Florida=int(input("Enter true or fale as 0 or 1 for Florida"))

inputs = [[RD, Admin, Market, California, Florida]]

result = model.predict(inputs)
print(result)
```
